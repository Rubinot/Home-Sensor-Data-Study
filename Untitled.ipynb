```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from statsmodels.tsa.stattools import adfuller

def load_and_preprocess(file_path):
    """Loads CSV data and processes the datetime components."""
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"The file {file_path} was not found.")
        
    print(f"[1/8] Loading and preprocessing dataset: '{file_path}'...")
    df = pd.read_csv(file_path)
    
    # Standardize column names (remove whitespace)
    df.columns = [col.strip() for col in df.columns]
    
    # Parse timestamp strings into Datetime objects
    df['Timestamp'] = pd.to_datetime(df['Timestamp'])
    df = df.sort_values('Timestamp').reset_index(drop=True)
    
    # Feature Engineering: Extract temporal components
    df['Hour'] = df['Timestamp'].dt.hour
    df['Date'] = df['Timestamp'].dt.date
    df['DayOfWeek'] = df['Timestamp'].dt.day_name()
    
    return df

def calculate_thermodynamics(df):
    """Calculates physical Dew Point temperature using the Magnus-Tetens formula."""
    print("[2/8] Calculating derived physical metrics (Dew Point)...")
    T = df['Temperature_C']
    RH = df['Humidity_Percent']
    
    # Constants for the Magnus-Tetens formula
    b = 17.625
    c = 243.04
    
    gamma = (b * T) / (c + T) + np.log(RH / 100.0)
    df['Dew_Point_C'] = (c * gamma) / (b - gamma)
    return df

def run_descriptive_stats(df):
    """Generates basic summary diagnostics."""
    print("\n==================================================")
    print("           DESCRIPTIVE METRICS SUMMARY            ")
    print("==================================================")
    print(df[['Temperature_C', 'Humidity_Percent', 'Dew_Point_C']].describe().to_string())

def run_correlation_tests(df):
    """Evaluates the relationship strength and linearity between variables."""
    print("\n==================================================")
    print("            RELATIONAL & CORRELATION TESTS        ")
    print("==================================================")
    
    p_coef, p_value = stats.pearsonr(df['Temperature_C'], df['Humidity_Percent'])
    s_coef, s_value = stats.spearmanr(df['Temperature_C'], df['Humidity_Percent'])
    
    print(f"Pearson Linear Correlation: {p_coef:8.4f}  (p-value: {p_value:.4e})")
    print(f"Spearman Rank Correlation : {s_coef:8.4f}  (p-value: {s_value:.4e})")
    
    if p_value < 0.05:
        print("-> The correlation is STATISTICALLY SIGNIFICANT (alpha = 0.05).")
    else:
        print("-> The correlation is NOT statistically significant.")

def run_normality_tests(df):
    """Tests if variables fit standard normal distributions using Shapiro-Wilk."""
    print("\n==================================================")
    print("        DISTRIBUTION PROFILING (NORMALITY)        ")
    print("==================================================")
    
    for col in ['Temperature_C', 'Humidity_Percent', 'Dew_Point_C']:
        stat, p_val = stats.shapiro(df[col].dropna())
        is_normal = "IS normal" if p_val > 0.05 else "is NOT normal"
        print(f"Shapiro-Wilk [{col:18}]: W={stat:.4f} | p-value={p_val:8.5f} | Data {is_normal}")

def detect_outliers_iqr(df):
    """Identifies observations deviating significantly from standard bounds."""
    print("\n==================================================")
    print("          OUTLIER DIAGNOSTICS (IQR METHOD)        ")
    print("==================================================")
    
    for col in ['Temperature_C', 'Humidity_Percent']:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        print(f"Variable '{col}': Found {len(outliers)} outlier(s) outside bounds [{lower_bound:.2f}, {upper_bound:.2f}]")
        if not outliers.empty:
            print(outliers[['Timestamp', 'Temperature_C', 'Humidity_Percent']].to_string(index=False))

def run_stationarity_tests(df):
    """Applies the Augmented Dickey-Fuller (ADF) test to evaluate time-series trends."""
    print("\n==================================================")
    print("          TIME-SERIES STATIONARITY TESTS          ")
    print("==================================================")
    
    for col in ['Temperature_C', 'Humidity_Percent']:
        try:
            result = adfuller(df[col].dropna())
            adf_stat, p_val = result[0], result[1]
            is_stationary = "STATIONARY (no long-term drift)" if p_val < 0.05 else "NON-STATIONARY (contains trend)"
            print(f"ADF Test [{col:18}]: Stat={adf_stat:8.4f} | p-value={p_val:8.5f} | Series is {is_stationary}")
        except Exception as e:
            print(f"ADF Test for {col} failed: {e}")

def run_temporal_profiling(df):
    """Groups data by hour and day of week to highlight recurring periodic patterns."""
    print("\n==================================================")
    print("          TEMPORAL PATTERN PROFILING             ")
    print("==================================================")
    
    print("\n--- Diurnal Behavior (Top 5 Hours by Mean Temperature) ---")
    hourly = df.groupby('Hour')[['Temperature_C', 'Humidity_Percent']].mean()
    print(hourly.sort_values(by='Temperature_C', ascending=False).head().to_string())
    
    print("\n--- Weekly Behavior (Mean Metrics Across Days of Week) ---")
    day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    weekly = df.groupby('DayOfWeek')[['Temperature_C', 'Humidity_Percent']].mean().reindex(day_order)
    print(weekly.to_string())

def generate_visualizations(df):
    """Creates explanatory plots and exports them as high-fidelity images."""
    print("\n[8/8] Generating diagnostic visualization figures...")
    sns.set_theme(style="whitegrid")
    
    # Plot 1: Time Series Trend Plot
    fig, ax1 = plt.subplots(figsize=(12, 6))
    color = 'tab:red'
    ax1.set_xlabel('Timeline / Date')
    ax1.set_ylabel('Temperature (°C)', color=color)
    ax1.plot(df['Timestamp'], df['Temperature_C'], color=color, alpha=0.6, linewidth=1.5, label='Temperature')
    ax1.tick_params(axis='y', labelcolor=color)
    
    ax2 = ax1.twinx()  
    color = 'tab:blue'
    ax2.set_ylabel('Relative Humidity (%)', color=color)
    ax2.plot(df['Timestamp'], df['Humidity_Percent'], color=color, alpha=0.6, linewidth=1.5, label='Humidity')
    ax2.tick_params(axis='y', labelcolor=color)
    
    plt.title('Environmental Sensor Metrics Timeline Trend Analysis', fontsize=14, pad=15)
    fig.tight_layout()
    plt.savefig('sensor_trends_report.png', dpi=200)
    plt.close()
    print(" -> Saved: 'sensor_trends_report.png'")
    
    # Plot 2: Correlation Scatter + Regression
    plt.figure(figsize=(8, 6))
    sns.regplot(data=df, x='Temperature_C', y='Humidity_Percent', 
                scatter_kws={'alpha':0.5, 'color':'teal'}, line_kws={'color':'crimson', 'linewidth':2})
    plt.title('Relational Interaction: Temperature vs. Relative Humidity', fontsize=12, pad=10)
    plt.xlabel('Temperature (°C)')
    plt.ylabel('Relative Humidity (%)')
    plt.tight_layout()
    plt.savefig('relational_interaction_chart.png', dpi=200)
    plt.close()
    print(" -> Saved: 'relational_interaction_chart.png'")

def main():
    # Provide the path to your CSV file
    csv_file = 'sensor_data.csv'
    
    try:
        # Pipeline Execution
        data = load_and_preprocess(csv_file)
        data = calculate_thermodynamics(data)
        
        run_descriptive_stats(data)
        run_correlation_tests(data)
        run_normality_tests(data)
        detect_outliers_iqr(data)
        run_stationarity_tests(data)
        run_temporal_profiling(data)
        generate_visualizations(data)
        
        print("\nPipeline execution complete! All structural information has been compiled.")
        
    except Exception as e:
        print(f"\n[ERROR] Pipeline execution failed: {e}")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd

def generate_daily_averages(input_file, output_file):
    # 1. Load the sensor dataset
    df = pd.read_csv(input_file)
    
    # 2. Convert the 'Timestamp' column to datetime format
    df['Timestamp'] = pd.to_datetime(df['Timestamp'])
    
    # 3. Extract just the date (YYYY-MM-DD) from the timestamp
    df['Date'] = df['Timestamp'].dt.date
    
    # 4. Group by 'Date' and calculate the mean for temperature and humidity
    daily_avg = df.groupby('Date')[['Temperature_C', 'Humidity_Percent']].mean().reset_index()
    
    # 5. Rename columns to indicate they represent daily averages
    daily_avg.columns = ['Date', 'Avg_Temperature_C', 'Avg_Humidity_Percent']
    
    # 6. Save the resulting summarized data to a new CSV file
    daily_avg.to_csv(output_file, index=False)
    print(f"Daily averages successfully saved to '{output_file}'")

# Execute the function
generate_daily_averages('sensor_data.csv', 'daily_average_sensor_data.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_sensor_trends(input_file, output_image_file):
    # 1. Load the daily average data
    df = pd.read_csv(input_file)
    
    # 2. Convert the 'Date' column to datetime format and sort chronologically
    df['Date'] = pd.to_datetime(df['Date'])
    df = df.sort_values('Date')
    
    # 3. Create subplots for Temperature and Humidity (sharing the same X-axis)
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True)
    
    # --- Plot 1: Temperature ---
    ax1.plot(df['Date'], df['Avg_Temperature_C'], color='tab:red', marker='o', linestyle='-', linewidth=2)
    ax1.set_ylabel('Average Temperature (°C)', color='tab:red', fontsize=12)
    ax1.tick_params(axis='y', labelcolor='tab:red')
    ax1.set_title('Daily Average Temperature and Humidity Trends', fontsize=14, pad=15)
    ax1.grid(True, linestyle='--', alpha=0.6)
    
    # --- Plot 2: Humidity ---
    ax2.plot(df['Date'], df['Avg_Humidity_Percent'], color='tab:blue', marker='s', linestyle='-', linewidth=2)
    ax2.set_xlabel('Date', fontsize=12)
    ax2.set_ylabel('Average Humidity (%)', color='tab:blue', fontsize=12)
    ax2.tick_params(axis='y', labelcolor='tab:blue')
    ax2.grid(True, linestyle='--', alpha=0.6)
    
    # 4. Format x-axis labels to avoid overlap and truncation
    plt.xticks(rotation=45)
    plt.tight_layout()
    
    # 5. Save the visualization as a high-quality image
    plt.savefig(output_image_file, dpi=300)
    print(f"Trend graph successfully saved as '{output_image_file}'")

# Execute the plotting function
plot_sensor_trends('daily_average_sensor_data.csv', 'temperature_humidity_trends.png')
```
